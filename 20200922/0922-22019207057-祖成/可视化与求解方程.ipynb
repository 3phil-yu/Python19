```python
#导入数据
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/Users/zucheng/Desktop/WorldIndex.csv')
df.head()
```

```python
#删除含缺失值的行
new_df=df.dropna(axis = 0) 
new_df
```

```python
#人均寿命直方图
plt.hist(x = new_df.Life_expectancy, # 指定绘图数据
         bins = 20, # 指定直方图中条块的个数
         color = 'steelblue', # 指定直方图的填充色
         edgecolor = 'black' # 指定直方图的边框色
         )
# 添加x轴和y轴标签
plt.xlabel('Life_expectancy')
plt.ylabel('frequency')
# 添加标题
plt.title('Mean age histogram')
# 显示图形
plt.show()
```

```python
#人均GDP直方图
plt.hist(x = new_df.GDP_per_capita, # 指定绘图数据
         bins = 20, # 指定直方图中条块的个数
         color = 'steelblue', # 指定直方图的填充色
         edgecolor = 'black' # 指定直方图的边框色
         )
# 添加x轴和y轴标签
plt.xlabel('GDP_per_capita')
plt.ylabel('frequency')
# 添加标题
plt.title('GDP histogram')
# 显示图形
plt.show()
```

```python
#人均寿命箱线图
plt.boxplot(x=new_df.Life_expectancy)
plt.ylabel("Life expectancy ")
plt.title("Life expectancy boxplot")
plt.show()
```

```python
#气泡图
#先定义气泡大小，rank 函数将大小列进行大小分配，越大的值分配结果也越高
#n 为倍数，用来调节气泡的大小，且看后头
size=new_df['Population'].rank()
n=3
colors = np.random.rand(len(new_df.Country)) 
plt.scatter(x=new_df.Life_expectancy,y=new_df.GDP_per_capita,s=size*n,alpha=0.6,c=colors)
plt.show()
```

```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root,fsolve

# 求解f(x)=2*sin(x)-x+1
rangex1 = np.linspace(-2,8)
rangey1_1,rangey1_2 = 2*np.sin(rangex1),rangex1-1
plt.figure(1)
plt.plot(rangex1,rangey1_1,'r',rangex1,rangey1_2,'b--')
plt.title('$2sin(x)$ and $x-1$')

def f1(x):
    return np.sin(x)*2-x+1

sol1_root = root(f1,[2])
sol1_fsolve = fsolve(f1,[2])
plt.scatter(sol1_fsolve,2*np.sin(sol1_fsolve),linewidths=9)
plt.show()

```

```python
# 求解线性方程组{3X1+2X2=3;X1-2X2=5}
def f2(x):
    return np.array([3*x[0]+2*x[1]-3,x[0]-2*x[1]-5])

sol2_root = root(f2,[0,0])
sol2_fsolve = fsolve(f2,[0,0])
print(sol2_fsolve) # [2. -1.5]

a = np.array([[3,2],[1,-2]])
b = np.array([3,5])
x = np.linalg.solve(a,b)
print(x)
```
