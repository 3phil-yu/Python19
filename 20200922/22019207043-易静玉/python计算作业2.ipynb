# #一.画图

```python
#导入绘图相关包
%matplotlib inline
import matplotlib as mpl
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt 
import pandas
import sympy
```

```python
#1.绘制所有洲人均寿命直方图

x=df.iloc[:,2]#截取文档第三列
x_bins = 30 #设置直方个数为30个
mu_life=np.mean(x)#人均期望寿命平均值为71.30127260524394
sigma_life=np.std(x)#人均期望寿命方差为8.485998724919199
n, bins, patches = plt.hist(x, x_bins,normed=1, color="gray",edgecolor="black", alpha=1)
y = mlab.normpdf(bins, mu_life, sigma_life)#拟合一条最佳正态分布曲线y 
plt.plot(bins, y, 'r--') #绘制y的曲线
plt.xlabel('Age',fontsize=14) #绘制x轴
plt.ylabel('Probability',fontsize=14) #绘制y轴
plt.title(r'Histogram of life expectancy: $\mu_life=71.30$,$\sigma_life=8.46$',fontsize=20)#中文标题
plt.subplots_adjust(left=0.15)#左边距为0.15
plt.show()
```

```python
#2.绘制所有洲人均GDP直方图

x=df.iloc[:,3]
x_bins = 30 
mu_GDP=np.mean(x)#人均GDP均值为12783.970323954876
sigma_GDP=np.std(x)#人均GDP方差为17911.987400863116
n, bins, patches = plt.hist(x, x_bins,normed=1, color="gray",edgecolor="black", alpha=1)
y= mlab.normpdf(bins, mu_GDP, sigma_GDP)#拟合一条最佳正态分布曲线y 
plt.plot(bins, y, 'r--') #绘制y的曲线
plt.xlabel('Daller',fontsize=14) #绘制x轴
plt.ylabel('Probability',fontsize=14) #绘制y轴
plt.title(r'Histogram of GDP_per_capita : $\mu_GDP=12783.97$,$\sigma_GDP=17911.99$',fontsize=20)#中文标题
plt.subplots_adjust(left=0.15)#左边距
plt.show()
```

```python
#3.人均寿命箱线图

#按各大洲名称重新划分文档，并重新命名为Af、As、Eu、No、Oc、So
Af=df.loc[df['Continent'].str.contains('Africa')]
As=df.loc[df['Continent'].str.contains('Asia')]
Eu=df.loc[df['Continent'].str.contains('Europe')]
No=df.loc[df['Continent'].str.contains('North America')]
Oc=df.loc[df['Continent'].str.contains('Oceania')]
So=df.loc[df['Continent'].str.contains('South America')]
Af_life=Af.iloc[:,2]
As_life=As.iloc[:,2]
Eu_life=As.iloc[:,2]
No_life=No.iloc[:,2]
Oc_life=Oc.iloc[:,2]
So_life=So.iloc[:,2]
#将各大洲国家人均寿命重新建立成一个列表
totaldata = {
    'Africa':Af_life.T ,
    'Asia':As_life.T,
    'Europe':Eu_life.T,
    "North America":No_life.T,
    'Oceania':Oc_life,
    'South America':So_life.T
}
df_box = pd.DataFrame(totaldata)#绘制箱型图
df_box.plot.box(title="Life_expectancy in each Continent",fontsize=8)#箱型图标题
plt.grid(linestyle="--", alpha=1)
plt.show()
```

```python
#4.绘制每个大洲的国家个数条形图

Af_rownumber=len(Af) #六大洲的行数等于各州的国家数量
As_rownumber=len(As)
Eu_rownumber=len(Eu)
No_rownumber=len(No)
Oc_rownumber=len(Oc)
So_rownumber=len(So)
#绘制各大洲国家数量条形图

#建立新列表（各大洲国家数目统计）
data_countrynumber = pd.Series([Af_rownumber, As_rownumber, Eu_rownumber, No_rownumber, Oc_rownumber, So_rownumber],index=['Africa','Asia','Europe','North America','Oceania','South America'])
axes = plt.subplots(1)
plt.xlabel('Country number',fontsize=14) #绘制x轴
plt.ylabel('Continent',fontsize=14) #绘制y轴
plt.title('Number of countries on each continent',fontsize=20)
data_countrynumber.plot.barh(ax=axes[1],color='gray',alpha=1)
plt.show()
```

```python
#5.绘制各大洲国家数量占比饼图

labels=['Africa','Asia','Europe','North America','Oceania','South America']
X=[Af_rownumber, As_rownumber, Eu_rownumber, No_rownumber, Oc__rownumber, So_rownumber]  
 
fig = plt.figure()
plt.pie(X,labels=labels,autopct='%1.2f%%') #画饼图（数据，数据对应的标签，百分数保留两位小数点）
plt.title("Pie chart of the number of countries in each continent",fontsize=20)
plt.show()  
```

```python
#6.绘制人均寿命和人均GDP散点图
x=df.iloc[:,2]
y=df.iloc[:,3]

plt.scatter(x, y, s=70)
 
# 设置图表标题并给坐标轴加上标签
plt.title('Scatter Plot of Life Expectancy and GDP per capita', fontsize=20)
plt.xlabel('Life Expectancy', fontsize=14)
plt.ylabel('GDP per capita', fontsize=14)
 
# 设置刻度标记的大小
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
#人均寿命和人均GDP呈指数关系增长
```

```python
#7.人均GDP与人均寿命的气泡图

#7.1 人均寿命与人口数量气泡图

data_LBC=df[['Continent','Population', 'Life_expectancy']] #将原表中大洲、人口、人均寿命列提取出来组成新的列表


size=data_LBC['Life_expectancy'].rank() #设置人均期望寿命气泡大小根据人口数量数据大小显示

plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Africa'],  #按各大洲国家人均期望寿命与人口数量成正比绘制气泡图,定义x、y、s
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Africa'],
           s =size[data_LBC.Continent == 'Africa'],
           color = 'steelblue', label = 'Africa', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Asia'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Asia'],
           s =size[data_LBC.Continent == 'Asia'],
           color = 'indianred', label = 'Asia', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Europe'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Europe'],
           s =size[data_LBC.Continent == 'Europe'],
           color = 'green', label = 'Asia', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Oceania'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Oceania'],
           s =size[data_LBC.Continent == 'Oceania'],
           color = 'black', label = 'Oceania', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'North America'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'North America'],
           s =size[data_LBC.Continent == 'North America'],
           color ='yellow', label = 'North America', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'South America'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'South America'],
           s =size[data_LBC.Continent == 'South America'],
           color = 'orange', label = 'South America', alpha =0.6
            )

# 添加x轴和y轴标签
plt.xlabel('Population',fontsize=14)
plt.ylabel('Life_expectancy',fontsize=14)
# 添加标题
plt.title('Bubble chart of population and life expectancy',fontsize=20)
# 添加图例
plt.legend()

# 显示图形
plt.show()
```
