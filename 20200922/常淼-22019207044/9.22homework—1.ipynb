```python
import numpy as np
import pandas as pd
data = pd.read_csv("c:\\users\\llcimi\\WorldIndex.csv",header=0)
```

```python
data.head(n=200)
df=data.dropna()
df
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt 
import pandas
import sympy
```

```python
#1.绘制所有洲人均寿命直方图
x=df.iloc[:,2]
x_bins = 30 
mu_life=np.mean(x)#71.30127260524394
sigma_life=np.std(x)#8.485998724919199
n, bins, patches = plt.hist(x, x_bins,normed=1, color="blue",edgecolor="black", alpha=1)
y = mlab.normpdf(bins, mu_life, sigma_life)#拟合一条最佳正态分布曲线y 
plt.plot(bins, y, 'r--') #绘制y的曲线
plt.xlabel('Age') #绘制x轴
plt.ylabel('Probability') #绘制y轴
plt.title(r'Histogram of life expectancy: $\mu_life=71.30$,$\sigma_life=8.46$')#中文标题
plt.subplots_adjust(left=0.15)#左边距
plt.show()
```

```python
#2.绘制所有洲人均GDP直方图
x=df.iloc[:,3]
x_bins = 30 
mu_GDP=np.mean(x)#12783.970323954876
sigma_GDP=np.std(x)#17911.987400863116
n, bins, patches = plt.hist(x, x_bins,normed=1, color="blue",edgecolor="black", alpha=1)
y= mlab.normpdf(bins, mu_GDP, sigma_GDP)#拟合一条最佳正态分布曲线y 
plt.plot(bins, y, 'r--') #绘制y的曲线
plt.xlabel('Daller') #绘制x轴
plt.ylabel('Probability') #绘制y轴
plt.title(r'Histogram of GDP_per_capita : $\mu_GDP=12783.97$,$\sigma_GDP=17911.99$')#中文标题
plt.subplots_adjust(left=0.15)#左边距
plt.show()
```

```python
#3.人均寿命箱线图
Af=df.loc[df['Continent'].str.contains('Africa')]
As=df.loc[df['Continent'].str.contains('Asia')]
Eu=df.loc[df['Continent'].str.contains('Europe')]
No=df.loc[df['Continent'].str.contains('North America')]
Oc=df.loc[df['Continent'].str.contains('Oceania')]
So=df.loc[df['Continent'].str.contains('South America')]
Af_life=Af.iloc[:,2]
As_life=As.iloc[:,2]
Eu_life=As.iloc[:,2]
No_life=No.iloc[:,2]
Oc_life=Oc.iloc[:,2]
So_life=So.iloc[:,2]
totaldata = {
    'Africa':Af_life.T ,
    'Asia':As_life.T,
    'Europe':Eu_life.T,
    "North America":No_life.T,
    'Oceania':Oc_life,
    'South America':So_life.T
}
df_box = pd.DataFrame(totaldata)
df_box.plot.box(title="Life_expectancy in each Continent")
plt.grid(linestyle="--", alpha=0.3)
plt.show()
```

```python
#4.绘制每个大洲的国家个数条形图
Af_rownumber=len(Af) #六大洲的行数等于各州的国家数量
As_rownumber=len(As)
Eu_rownumber=len(Eu)
No_rownumber=len(No)
Oc__rownumber=len(Oc)
So_rownumber=len(So)
#绘制各大洲国家数量条形图
data_countrynumber = pd.Series([Af_rownumber, As_rownumber, Eu_rownumber, No_rownumber, Oc__rownumber, So_rownumber],index=['Africa','Asia','Europe','North America','Oceania','South America'])
axes = plt.subplots(1)
plt.xlabel('Country number') #绘制x轴
plt.ylabel('Continent') #绘制y轴
plt.title('Number of countries on each continent')
data_countrynumber.plot.barh(ax=axes[1],color='b',alpha=1)
```

```python
#5.绘制各大洲国家数量占比饼图
labels=['Africa','Asia','Europe','North America','Oceania','South America']
X=[Af_rownumber, As_rownumber, Eu_rownumber, No_rownumber, Oc__rownumber, So_rownumber]  
 
fig = plt.figure()
plt.pie(X,labels=labels,autopct='%1.2f%%') #画饼图（数据，数据对应的标签，百分数保留两位小数点）
plt.title("Pie chart of the number of countries in each continent")
plt.show()  
```

```python
#6.绘制人均寿命和人均GDP散点图
x=df.iloc[:,2]
y=df.iloc[:,3]

plt.scatter(x, y, s=50)
 
# 设置图表标题并给坐标轴加上标签
plt.title('Scatter Plot of Life Expectancy and GDP per capita', fontsize=20)
plt.xlabel('Life Expectancy', fontsize=14)
plt.ylabel('GDP per capita', fontsize=14)
 
# 设置刻度标记的大小
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```

```python
#7.人均GDP与人均寿命的气泡图
#7.1 人均寿命与人口数量气泡图

data_LBC=df[['Continent','Population', 'Life_expectancy']]
data_GBC=df[['Continent','Population', 'GDP_per_capita']]

size=data_LBC['Life_expectancy'].rank()

plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Africa'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Africa'],
           s =size[data_LBC.Continent == 'Africa'],
           color = 'steelblue', label = 'Africa', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Asia'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Asia'],
           s =size[data_LBC.Continent == 'Asia'],
           color = 'indianred', label = 'Asia', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Europe'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Europe'],
           s =size[data_LBC.Continent == 'Europe'],
           color = 'green', label = 'Asia', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'Oceania'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'Oceania'],
           s =size[data_LBC.Continent == 'Oceania'],
           color = 'black', label = 'Oceania', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'North America'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'North America'],
           s =size[data_LBC.Continent == 'North America'],
           color ='yellow', label = 'North America', alpha = 0.6
            )
plt.scatter(x = data_LBC.Population[data_LBC.Continent == 'South America'],
           y = data_LBC.Life_expectancy[data_LBC.Continent == 'South America'],
           s =size[data_LBC.Continent == 'South America'],
           color = 'orange', label = 'South America', alpha = 0.6
            )

# 添加x轴和y轴标签
plt.xlabel('Population')
plt.ylabel('Life_expectancy')
# 添加标题
plt.title('Bubble chart of population and life expectancy')
# 添加图例
plt.legend()

# 显示图形
plt.show()
```
