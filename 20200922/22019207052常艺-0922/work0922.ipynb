```python
data2=data.dropna(subset=['Life_expectancy'])
data2.info()
```

```python
import matplotlib.pyplot as plt
x=data2['Life_expectancy']
plt.rcParams['font.sans-serif']=['SimHei']
plt.xlabel('平均寿命')
plt.ylabel('数量')
plt.title("人均寿命分布直方图")
plt.hist(x,bins=[50,60,70,80,90,100],facecolor='pink',edgecolor='black',alpha=0.7)
plt.show()
```

```python
x=data2['Life_expectancy']
plt.boxplot(x,whis=1.5,widths=0.3,patch_artist=True,showmeans=True,boxprops={'facecolor':'RoyalBlue'},flierprops={'markerfacecolor':'red','markeredgecolor':'red','markersize':3},meanprops={'marker':'h','markerfacecolor':'black','markersize':8},medianprops={'linestyle':'--','color':'orange'},labels=[''])
plt.title("人均寿命分布箱型图")
plt.show()
```

```python
data3=data.dropna(subset=['GDP_per_capita'])
data3.info()
```

```python
import matplotlib.pyplot as plt
x=data3['GDP_per_capita']
plt.rcParams['font.sans-serif']=['SimHei']
plt.xlabel('人均GDP')
plt.ylabel('数量')
plt.title("人均GDP分布直方图")
plt.hist(x,bins=[0,5000,10000,15000,20000,25000,30000,35000,40000,50000,60000,70000,80000,90000],facecolor='darksalmon',edgecolor='black',alpha=0.7)
plt.show()
```

```python
data4=data3.dropna(subset=['Life_expectancy'])
data4.info()
```

```python
x=data4['GDP_per_capita']
y=data4['Life_expectancy']
plt.scatter(x[:48],y[:48],c='pink',marker='o')
plt.scatter(x[48:86],y[48:86],c='royalblue',marker='v')
plt.scatter(x[86:125],y[86:125],c='darkcyan',marker='*')
plt.scatter(x[125:144],y[125:144],c='deeppink',marker='8')
plt.scatter(x[144:153],y[144:153],c='lightgrey',marker='s')
plt.scatter(x[153:],y[153:],c='darkorange',marker='p')
plt.title("人均寿命与人均GDP分布散点图")
plt.show()
```

```python
import seaborn as sns
def DrawBubble(read_name):#气泡图
    sns.set(style = "whitegrid")#设置样式
    x = data4.GDP_per_capita#X轴数据
    y = data4.Life_expectancy#Y轴数据
    z = data4.Population#用来调整各个点的大小s
    plt.rcParams['font.sans-serif']=['SimHei']
    cm = plt.cm.get_cmap('RdYlBu')
    fig,ax = plt.subplots(figsize = (12,10))
    bubble = ax.scatter(x,y, s = z*0.00001, c = z, cmap = cm, linewidth = 0.5, alpha = 0.5)
    ax.grid()
    fig.colorbar(bubble)
    ax.set_xlabel('人均GDP', fontsize = 15)#X轴标签
    ax.set_ylabel('人均寿命', fontsize = 15)#Y轴标签
    plt.title("人均寿命与人均GDP分布气泡图")
plt.show()
if __name__=='__main__':
    DrawBubble("WorldIndex.csv")#气泡图
```
