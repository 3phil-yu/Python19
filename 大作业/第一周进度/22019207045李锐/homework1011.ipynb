```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
filepath1="E:\\homework\\NumericalPython\\Basket-2018-06-01-2018-07-05.csv"
Basket=pd.read_csv(filepath1)
Basket.head()
```

```python
Basket.info()
```

```python
Basket_time_sum = Basket.groupby('SHOP_DATETIME')['BASK_SPEND'].sum()
Basket_time_sum
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #正常显示中文标签 
Basket_time_sum.plot(kind='bar',color="blue",title="每日营业额",figsize=(14,8))
```

```python
Basket_code_sum = Basket.groupby('STORE_CODE')['BASK_SPEND'].sum()
Basket_code_sum
```

```python
plt.rcParams['font.sans-serif']=['SimHei'] #正常显示中文标签 
Basket_code_sum.plot(kind='bar',color="blue",title="商铺营业额",figsize=(14,8))
```

```python
Basket_bc_sum = Basket.groupby('BASK_CODE')['BASK_SPEND'].sum()
Basket_bc_sum
```

```python
Basket_pos_sum = Basket.groupby('POS_NO')['BASK_SPEND'].sum()
Basket_pos_sum
```
